b) Napisati Python program koji za početnu vrijednost n računa n-ti Fibonacci broj. Obavezno koristiti for petlju

```python
def Fibonacci(n):
    if n<= 0:
        print("Incorrect input")
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return Fibonacci(n-1)+Fibonacci(n-2)

n = int(input("Enter the value for calculating the n'th Fibonacci number: "))
print(Fibonacci(n))
```

Definisati funkciju fact(n) koja računa faktorijel broja n. Zatim definisati funkciju choose(n, k) koja
računa (nCk)

```python
def fact(n):
    res = 1
    for i in range(2, n+1):
        res *= i
    return res

def choose(n,k):
    res = fact(n) / (fact(k) * fact(n - k))
    return res

n = int(input("Enter n: "))
k = int(input("Enter k: "))
print("Result: ", choose(n,k))
```

g) Definisati funkciju kvadratna(M) koja prima NumPy 1D niz, te isti preoblikuje u kvadratnu matricu, ukoliko
je to moguće. Ukoliko nije moguće, funkcija ne treba da uradi ništa.

```python
def kvadratna(M):
    length = len(M)
    sqrt_length = int(np.sqrt(length))

    if sqrt_length * sqrt_length != length:
        return

    square_matrix = np.reshape(M, (sqrt_length, sqrt_length))
    print(square_matrix)

nums = []
length = int(input("Enter list length: "))
for i in range(length):
  val = int(input("Input " + str(i + 1) + ". element of the list: "))
  nums.append(val)

kvadratna(nums)
```

# Zadatak 4


a) Neka je data funkcija x = sin(n) na 100 tačaka (n = 1:100). Iscrtati linijski dijagram ove funkcije. Postaviti odgovarajuće labele, te naslov. Zatim preko tog dijagrama (na isti graf) nacrtati funkciju y = cos(n), također na 100 uzoraka. Uključiti mrežu (grid)

```python
import matplotlib.pyplot as plt

n = np.arange(1, 101)
x = np.sin(n)
y = np.cos(n)

plt.plot(n, x, label='x = sin(n)')
plt.plot(n, y, label='y = cos(n)')
plt.xlabel('n')
plt.ylabel('Vrijednost')
plt.title('Grafici funkcija x = sin(n) i y = cos(n)')
plt.grid(True)
plt.legend()
plt.show()
```

b) Ponoviti prethodni zadataka, ali sada iscrtati u istom prozoru (ali na različitim graficima) signale x i y. Postaviti odgovarajuće labele i naslove za oba grafika.

```python
n = np.arange(1, 101)
x = np.sin(n)
y = np.cos(n)

plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.plot(n, x, label='x = sin(n)', color='blue')
plt.xlabel('n')
plt.ylabel('Vrednost')
plt.title('Grafik funkcije x = sin(n)')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(n, y, label='y = cos(n)', color='red')
plt.xlabel('n')
plt.ylabel('Vrednost')
plt.title('Grafik funkcije y = cos(n)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
